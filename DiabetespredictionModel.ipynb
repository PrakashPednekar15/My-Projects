```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as ttp
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
db_data=pd.read_csv('/content/drive/MyDrive/Machine learning datasets/diabetes.csv')
```

```python
#splitting data into test
X=db_data.drop(columns='Outcome',axis=1)
Y=db_data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
#data standardization
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
std_data=scaler.transform(X)
```

```python
print(std_data)
```

```python
X=std_data
Y=db_data['Outcome']
```

```python
print(X)
print(Y)
```

```python
#splitting data into training data and test data
X_train,X_test,Y_train,Y_test=ttp(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train,X_test.shape)
```

```python
clasifier= svm.SVC(kernel='linear')
```

```python
clasifier.fit(X_train,Y_train)
```

```python
#model evaluation
#checking accuracy on train data
X_train_pred=clasifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,Y_train)
```

```python
print("accuracy score ofthe training data:",training_data_accuracy)
```

```python
#checking accuracy of test data
X_test_pred=clasifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,Y_test)
```

```python
print("accuracy score of the test data:",test_data_accuracy)
```

```python
#building a predictive system
#Pregnancies	Glucose	BloodPressure	SkinThickness	Insulin	BMI	DiabetesPedigreeFunction	Age
pregnancy=int(input("enter the pregnency count"))
glucose_level=int(input("enter glucose level "))
bloodpressure=int(input("enter bloodpressure "))
skinthickness=int(input("enter skinthickness "))
insulin=int(input("enter the insulin range "))
bmi=float(input("enter the bmi "))
dpf=float(input("enter the diabetes pedigreefunction  "))
age=int(input("enter your age "))

input_data=[pregnancy,glucose_level,bloodpressure,skinthickness,insulin,bmi,dpf,age ]

input_data_array=np.asarray(input_data)
input_data_reshape=input_data_array.reshape(1,-1)

#standardizing the input data
std_data1=scaler.transform(input_data_reshape)
#print(std_data)

prediction=clasifier.predict(std_data1)
print(prediction)

if(prediction[0]==0):
  print("the person is non diabetic")
else:
  print("the person is diabetic")
```
